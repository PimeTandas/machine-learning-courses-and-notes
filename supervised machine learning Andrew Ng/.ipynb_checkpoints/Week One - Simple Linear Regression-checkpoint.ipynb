## Modeling
The model that maps x to y is linear regression. We are going to start by usinng some predefined (perfect) values for the weight w and bias b. 

w = 100
b = 100

```python
w = 200
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

```python
# Simple Linear Regression using a loop:

def using_for_loop(X, w, b):
    """
    Computes a prediction yHat using X and weights w and bias b.
    Arguments:
     - X - The training features X_train, length m.
     - w - The weight.
     - b - The bias.
     Returns:
     - yHat - A array of predicted values same length as X (m).
    """
    m = X.shape[0]
    yHat = np.zeros(m)
    for i in range(m):
        yHat[i] = X[i] * w + b
        
    return yHat
```

```python
yHat = using_for_loop(X_train, w, b)
print(yHat)
```

```python
plt.plot(X_train, yHat)
plt.scatter(X_train, y_train, marker = 'x', color = 'red')
```

## Computing Cost
Cost is a measurement of how well our model is performing, at predicting values. We will be using ordienry least square (OLS) which takes the average of the squared differences (differences being yPredict - yActual).

$$ Cost = \frac{1}{2m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)})^2 $$


```python
def compute_cost(X, w, b, y):
    """
    Calculates the cost of the linear regression model.
    Arguments:
     - X - The training features X_train, length m.
     - w - The weight.
     - b - The bias.
     - y - The actual y values assoicated with our input values X.
     Returns:
     - total_cost - The toal (error) cost of our function for the values we have used for w and b. 
    """
    m = X_train.shape[0]
    
    cost_sum = 0
    for i in range(m):
        yHat = X[i] * w + b
        cost = (yHat - y[i]) ** 2
        cost_sum += cost
    total_cost = cost_sum / (2 * m)
    
    return total_cost
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480,  430,   630, 730,])

yHat = using_for_loop(x_train, w, b)
cost = compute_cost(x_train, w, b, y_train)
print(f"Cost for this simple model is : {cost}")
```

## Computing the Gradient
In linear regression we aim to utilze the input data by fitting it to paramaters which minimise the error over the whole dataset.
When the cost function is:

$$ Cost = \frac{1}{2m} \sum_{i=1}^{m} yHat - y^{(i)})^2 $$
\
The gradient of the cost function for w and b can be found using:
$$ \frac{\partial J(\theta)}{\partial w} = \frac{1}{m} \sum_{i=1}^{m} yHat - y^{(i)}) \cdot x^{(i)} $$

$$ \frac{\partial J(\theta)}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} yHat - y^{(i)}) $$

```python
def compute_gradient(x, y, w, b):
    """
    Compute the gradient for linear regression.
    Arguments:
     - X - The training features X_train, length m.
     - y - The actual y values assoicated with our input values X.
     - w - The weight.
     - b - The bias.
    Returns:
     - gw - Gradient for paramater w.
     - gb - Gradient for paramater b.
    """
    m = x.shape[0]
    
    gwSum = 0
    gbSum = 0
    
    for i in range(m):
        yHat = w * x[i] + b
        gw = (yHat - y[i]) * x[i]
        gb = yHat - y[i]
        gwSum +=gw
        gbSum +=gb
    gw = gwSum / m
    gb = gbSum / m
    
    return gw, gb
```

## Gradient Descent
The final thing we shall do in this notebook is implement gradient descent. To do this we will need:
- <b>The function to compute gradient.</b>
- <b>A function Compute the cost.</b>
- <b>A function to compute dradient descent using the two functions above.</b> We will implement this below, using the following equation:

$$ w = w - \alpha \cdot \frac{1}{m} \sum_{i=1}^{m} yHat - y^{(i)}) \cdot x^{(i)} $$
      $$ b = b - \alpha \cdot \frac{1}{m} \sum_{i=1}^{m} yHat - y^{(i)}) $$

```python
def gradient_descent(X, y, wInit, bInit, alpha, num_iter, cost_function, gradient_function):
    """
    Performs gradient descent to find the optimal values for w and b to best fit a line to the training dataset.
    Arguments:
     - X - The training features X_train, length m.
     - y - The true values (dependent variable) when given input values X.
     - w - The weight.
     - b - The Bias.
     - alpha - Learning rate.
     - num_iter - How many itterations we want to allow gradient descent to run.
     - cost_function - The function used to calculate cost.
     - gradient_function - The function used to calculate gradient.
    Returns:
     - w_opt - Optimal w value.
     - b_opt - Optimal b value.
     - cost_history - The history of the cost as we perform gradient descent. (allows us to see if cost is decreasing).
    """
    cost_history = []
    b = bInit
    w = wInit
    
    for i in range(num_iter):
        # Work out the graident given the current values of w and b.
        gw, gb = compute_gradient(X, y, w, b)
        
        b = b - alpha * gb 
        w = w - alpha * gw
        
        if i < 1000000:
            cost_history.append(compute_cost(X, w, b, y))
    
    return w, b 
```

```python
# Tesing out everything so far.

# Training data.
X_train = np.array([1, 2])
y_train = np.array([300, 500])

# Paramaters set to zero for w and b.
w_init = 0
b_init = 0

# Some random gradient descent settings.

num_iter = 1000000
alpha = 1.0e-2

# Run gradient descent.
w_final, b_final = gradient_descent(X_train, y_train, w_init, b_init, alpha, num_iter, compute_cost, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final},{b_final})")
```
