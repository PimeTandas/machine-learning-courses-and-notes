# Polynomial Regression

Polynomial Regression is a non linear regression model used tackle non linear data. Often a linear regression model/ line will not fit the data very well, we thus need polynomial regression inorder to fit a curve to a graph.
\
\
We will be looking at a simple senario where a HR department has collected the data found in the csv file: Salary_Data.csv.

## Import Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

## Importing the Data

```python
df = pd.read_csv('Datasets/Position_Salaries.csv')
X = df.iloc[:, 1:-1].values 
y = df.iloc[:, -1].values
```

## Training a Linear Regression Model

```python
myLR = LinearRegression()
myLR.fit(X, y)
```

## Training a Polynomial Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)

myLRPoly = LinearRegression()
myLRPoly.fit(X_poly, y)
```

## Visualising Linear Regression Model Results

```python
plt.scatter(X, y, marker = 'x', color='red')
plt.plot(X, myLR.predict(X))
plt.title('Linear Model Fitted to non-Linear data')
plt.xlabel('Expereince in years')
plt.ylabel('Salary 1000s £')
```

## Visualising Polynomial Regression Model Results

```python
plt.scatter(X, y, marker = 'x', color='red')
plt.plot(X, myLRPoly.predict(X_poly))
plt.title('Polynomial Model Properly Fitted')
plt.xlabel('Expereince in years')
plt.ylabel('Salary 1000s £')
```

```python
# Repeating above but with 4 degrees.
poly_reg2 = PolynomialFeatures(degree = 4)
X_poly2 = poly_reg2.fit_transform(X)
myLRPoly2 = LinearRegression()
myLRPoly2.fit(X_poly2, y)


plt.scatter(X, y, marker = 'x', color='red')
plt.plot(X, myLRPoly2.predict(X_poly2))
plt.title('Polynomial Model Properly Fitted')
plt.xlabel('Expereince in years')
plt.ylabel('Salary 1000s £')
```

```python
# Same graph but attempting to smoothen it out.
X_data = np.arange(min(X), max(X), 0.1)
X_data = X_data.reshape(-1, 1)
plt.scatter(X, y)
plt.plot(X_data, myLRPoly2.predict(poly_reg2.fit_transform(X_data)))
plt.title('Polynomial Model Properly Fitted')
plt.xlabel('Expereince in years')
plt.ylabel('Salary 1000s £')
```

## Making Predictions with Linear Model

```python
myLR.predict([[6.5]])
```

## Making Predictions with Polynomal Model

```python
myLRPoly.predict(poly_reg2.fit_transform([[6.5]]))
```
