# R-Squared
We know that for each data point in our model we can look at the sum of residuals squared, infact we build our regresson line by minimising this sum. The residual sum of squares is:
$$ RSS (SSres) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

To caculate $R^2$ we also need to cauclate the total sum squared:

$$ TSS (SStot) = \sum_{i=1}^{n} (y_i - y_{avr})^2 $$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4, 5])
y = np.array([2.3, 4.8, 7.2, 9.7, 11.5])
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(X, y)
plt.plot(X, (X * 2) + 1)
plt.title("Residual Sum Square")

plt.subplot(1, 2, 2)
plt.scatter(X, y)
plt.plot(X, (X*0 + 7))
plt.title("Total Sum Square (Avrage)")
```

To calculate $R^2$:
$$R^2 = 1 \frac{SSres}{SStot}$$

We we can see that as SSres decreases then the $R^2$ value also decreses, and thus increases and punishes the $R^2$ when the residual sum is large, particulaly when it is larger than the Avrage sum squared.

#### *Rule of Thumb*
If the $R^2$ Value is:
- 1.0 then the model is a perfect fit (very unlikely).
- 0.9 then the model is very very good. 
- less than 0.7 then its not good not bad.
- less than 0.4 model is rubbish and no good for the data.

## Use of $R^2$
 - <b>Model Acessment</b> - Used to determine the goodness of fit of a regresson model. 
 - <b>Compairing Models</b> - Used to compair two or more models, the model with the larger $R^2$ number should be considered as "performing better".
 - <b>Predictive Power</b> - It gives an idea of how well the independent variables explain the variability of the dependent variable.
